### Incluindo Base line

```python
import numpy as np
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
         'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

pontuacao = np.array([16,13,15,12,10,20,25,40,30,25,60,50]) / 100
pontuacao2 = np.array([33,30,40,35,30,25,10,60,45,40,50,30]) / 100
pontuacao3 = np.array([39,42,37,37,40,35,28,45,45,50,55,40]) / 100

plt.plot(meses, pontuacao, c='red', ls='--', marker='o', label='Decision Tree')
plt.plot(meses, pontuacao2, marker='^', label='K-NN')
plt.plot(meses, pontuacao3, c='black', marker='s', label='Naive Bayes')

baseline_mensal = [max(pontuacao[i], pontuacao2[i], pontuacao3[i]) for i in range(12)]
plt.scatter(meses, baseline_mensal, color='green', s=80, label='Baseline mensal (máx)', zorder=5)

plt.legend(loc='lower right')
plt.title('Avaliação de desempenho de modelos em 2019')
plt.ylabel('Acurácia')
plt.ylim(0.0, 1.0)
plt.grid(True)
plt.show()
```

```python
models = [ "Decision Tree", 'K-NN', 'Naive Bayes']
x= meses
y = [ pontuacao, pontuacao2, pontuacao3]

plt.margins(0,0)
plt.stackplot(x,y, labels=models, alpha=0.7)
plt.legend(loc='upper left')
plt.title('Gráfico de áreas emplicadas')
plt.show()
```
